```python
import pandas as pd

# Load the red and white wine data
red_wine = pd.read_csv('winequality-red.csv', delimiter=';')
white_wine = pd.read_csv('winequality-white.csv', delimiter=';')
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Plot histogram for red wine quality distribution
sns.histplot(red_wine['quality'], kde=True)
plt.title('Distribution of Red Wine Quality')
plt.xlabel('Wine Quality')
plt.ylabel('Frequency')
plt.show()
```

```python
# Plot histogram for white wine quality distribution
sns.histplot(white_wine['quality'], kde=True)
plt.title('Distribution of White Wine Quality')
plt.xlabel('Wine Quality')
plt.ylabel('Frequency')
plt.show()
```

```python
# Boxplot for alcohol content vs wine quality (Red Wine)
sns.boxplot(x='quality', y='alcohol', data=red_wine)
plt.title('Alcohol vs Quality (Red Wine)')
plt.xlabel('Wine Quality')
plt.ylabel('Alcohol Content')
plt.show()
```

```python
# Boxplot for alcohol content vs wine quality (White Wine)
sns.boxplot(x='quality', y='alcohol', data=white_wine)
plt.title('Alcohol vs Quality (White Wine)')
plt.xlabel('Wine Quality')
plt.ylabel('Alcohol Content')
plt.show()
```

```python
# Correlation matrix for red wine dataset
corr_matrix_red = red_wine.corr()
plt.figure(figsize=(10,8))
sns.heatmap(corr_matrix_red, annot=True, cmap='coolwarm', fmt='.2f')
plt.title('Correlation Matrix - Red Wine')
plt.show()
```

```python
# Correlation matrix for white wine dataset
corr_matrix_white = white_wine.corr()
plt.figure(figsize=(10,8))
sns.heatmap(corr_matrix_white, annot=True, cmap='coolwarm', fmt='.2f')
plt.title('Correlation Matrix - White Wine')
plt.show()
```

```python
from sklearn.ensemble import RandomForestRegressor

# Define features and target variable for red wine
X_red = red_wine.drop('quality', axis=1)
y_red = red_wine['quality']

# Random Forest Model
rf_model = RandomForestRegressor(n_estimators=100, random_state=42)
rf_model.fit(X_red, y_red)

# Get feature importance
importances = rf_model.feature_importances_

# Plot feature importances
plt.figure(figsize=(10,6))
sns.barplot(x=importances, y=X_red.columns)
plt.title('Feature Importance - Random Forest (Red Wine)')
plt.xlabel('Importance')
plt.ylabel('Features')
plt.show()
```

```python
# Define features and target variable for white wine
X_white = white_wine.drop('quality', axis=1)
y_white = white_wine['quality']

# Random Forest Model for white wine
rf_model_white = RandomForestRegressor(n_estimators=100, random_state=42)
rf_model_white.fit(X_white, y_white)

# Get feature importance
importances_white = rf_model_white.feature_importances_

# Plot feature importances
plt.figure(figsize=(10,6))
sns.barplot(x=importances_white, y=X_white.columns)
plt.title('Feature Importance - Random Forest (White Wine)')
plt.xlabel('Importance')
plt.ylabel('Features')
plt.show()
```
